# Task 22

```python
l=[]
for i in range(100):
    df=ds['damageDealt'].sample(50)
    me=round(df.mean(),2)    
    l.append(me)
binn=np.arange(70,180,1)
plt.hist(l,binn)
plt.show()
sns.distplot(l,binn)
```

Average damage dealt is around 125 for a sample size of 50 taken 100 times
From this we can say that in general a player takes a damage of 125 and above for all match types
